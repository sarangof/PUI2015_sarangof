+ Generate 100 samples of different sizes N (N>10 & N<2000) from each of 6 different distributions (500 samples in total), all with the same population mean.
Include a Normal, a Poisson, a Binomial, a Chi-Squared distribution, and 2 more of your choice.


```python
import numpy as np
%matplotlib inline
import pylab as pl
import matplotlib.pyplot as plt
```

```python
sizes = np.array(np.arange(10,2000,20)) #Generate 100 sizes
popMean = 50
dist={}
dist['Normal']={}
dist['Poisson']={}
dist['Chisquare']={}
dist['Uniform']={}
dist['Exponential']={}
for i in range(0,99):
    dist['Normal'][i]=np.random.normal(popMean,4,sizes[i])
    dist['Poisson'][i]=np.random.poisson(popMean,sizes[i])
    dist['Chisquare'][i]=np.random.chisquare(popMean,sizes[i])
    dist['Uniform'][i]=np.random.uniform(popMean-popMean/2,popMean+popMean/2,sizes[i])
    dist['Exponential'][i]=np.random.exponential(popMean,sizes[i])
```

+ For each sample plot the sample mean against the sample size N (if you want you can do it with the sample standard deviation as well).

```python
points = {}
points['Normal']={}
points['Poisson']={}
points['Chisquare']={}
points['Uniform']={}
points['Exponential']={}
fig_mu_N = pl.figure(figsize=(10,10))
meanplot = fig_mu_N.add_subplot(111)
for d in dist: 
    for i in dist[d]:        
        points[d][i]=dist[d][i].mean()
        meanplot.plot(i, points[d][i], 'o')  
```

+ PLOT the distributions of all sample means (together for all distributions). 

Mandatory: as a histogram,
optional: in any other way you think is convincing

```python
lst = np.zeros(490050)#588060)
cont = 0
for d in dist:
    for i in dist[d]:
        for j in dist[d][i]:
            lst[cont] =  j
            cont+=1
plt.hist(lst,bins=100)
plt.show()
```
